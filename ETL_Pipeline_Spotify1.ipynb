```python
#Now we have to create it for all the entire 50 records
album_list = []
for row in data['items']:
  album_id = row['track']['album']['id']
  album_name = row['track']['album']['name']
  album_release_date = row['track']['album']['release_date']
  album_total_tracks = row['track']['album']['total_tracks']
  album_url = row['track']['album']['external_urls']['spotify']
  album_element = {'album_id': album_id, 'album_name': album_name, 'album_release_date':album_release_date,'album_total_tracks':album_total_tracks, 'album_url':album_url}
  album_list.append(album_element)   #append this element inside albumn element
```

```python
'''Now we need to create it for artist datat - for transaforming the extracted data - data['items'][0]
data['items'][0]['track']['artists'] # we need to get relevant information from this - so we need to iterate over data items - song id to relate
each data with it - by getting song id for all the songs'''
data['items'][0]['track']['artists']
```

```python
#create list with artists information
artists_list =[]
for row in data['items']:
  song_id = row['track']['id']
  for artists in row ['track']['artists']:
    artists_element = {
        'artists_id':artists['id'] ,'song_id': song_id, 'artist_name':artists['name'],'external_url':artists['href']
    }
    artists_list.append(artists_element)
```

```python
#similarly we need to do it for songsblock - to extract element items
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {
        'album_id':album_id,
        'song_id':song_id,
        'song_name':song_name,
        'song_duration_':song_duration,
        'song_url':song_url,
        'song_popularity':song_popularity,
        'song_added':song_added,
        'artist_id':artist_id
    }
    song_list.append(song_element)
```

```python
#we have it all list of dictionaries/object - it need to be converted into table format using pandas
import pandas as pd

album_df = pd.DataFrame.from_dict(album_list) #converts object into data frame album_df.head()
```
